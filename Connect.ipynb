```python
# CONNECT MUNICIPIO

MUNICIPIO = '1MToNUrBmVft-B_WaSi9Vz_6LG1QDP_Zk'

# URL direta para download
url_MUN = f'https://drive.google.com/uc?export=download&id={MUNICIPIO}'

# Fazendo o download do conteúdo
response = requests.get(url_MUN)
Data_MUN = StringIO(response.text)

# Lendo com pandas
mun = pd.read_csv(Data_MUN, encoding='utf-8', sep=';')

mun = mun.drop('NO_MUN_MIN', axis=1)

display(mun.head())
```
